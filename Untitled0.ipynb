```python
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
from google.colab import drive

from darts import TimeSeries
from darts.models import TFTModel, LightGBMModel
from darts.dataprocessing.transformers import Scaler, StaticCovariatesTransformer

# ========== 1. 挂载与读取 ==========
drive.mount('/content/drive', force_remount=True)

path = '/content/drive/MyDrive/carbon-project/建模面板数据.csv'
df = pd.read_csv(path, encoding='gbk')

df.columns = df.columns.str.strip()
df = df.sort_values(['Province', 'Year']).reset_index(drop=True)
df['Year_dt'] = pd.to_datetime(df['Year'].astype(int), format='%Y')

target_col = 'carbon_intensity'

# 优先使用更多变量，提高解释力
candidate_cov_cols = [
    'Coal_Share',
    'Gas_Share',
    'Oil_Share',
    'Clean_Share',
    'ln_pgdp',
    'urban_rate',
    'second_share',
    'ln_rd',
    'env_invest'
]

cov_cols = [c for c in candidate_cov_cols if c in df.columns]

print(">>> 本次使用的协变量：")
print(cov_cols)

df = df.dropna(subset=['Province', 'Year_dt', target_col] + cov_cols).copy()

# ========== 2. 预测未来协变量：2023-2026 ==========
future_cov_list = []

for col in cov_cols:
    print(f">>> 正在预测未来协变量：{col}")

    cov_series = TimeSeries.from_group_dataframe(
        df,
        group_cols='Province',
        time_col='Year_dt',
        value_cols=col,
        freq='YS',
        fill_missing_dates=True
    )

    scaler_tmp = Scaler()
    cov_scaled_tmp = scaler_tmp.fit_transform(cov_series)

    model_cov = LightGBMModel(
        lags=5,
        output_chunk_length=4,
        random_state=42,
        use_static_covariates=False
    )

    model_cov.fit(cov_scaled_tmp)
    pred_scaled = model_cov.predict(n=4, series=cov_scaled_tmp)
    pred = scaler_tmp.inverse_transform(pred_scaled)

    temp_list = []

    for i, s in enumerate(pred):
        province_name = cov_series[i].static_covariates['Province'].iloc[0]

        temp = s.to_dataframe().reset_index()
        temp.columns = ['Year_dt', col]
        temp['Year'] = temp['Year_dt'].dt.year
        temp['Province'] = province_name

        temp_list.append(temp[['Province', 'Year', col]])

    future_cov_list.append(pd.concat(temp_list, ignore_index=True))

future_cov_df = future_cov_list[0]

for item in future_cov_list[1:]:
    future_cov_df = future_cov_df.merge(item, on=['Province', 'Year'], how='left')

# ========== 3. 能源比例修正 ==========
energy_share_cols = [c for c in ['Coal_Share', 'Gas_Share', 'Oil_Share', 'Clean_Share'] if c in future_cov_df.columns]

for col in energy_share_cols:
    future_cov_df[col] = future_cov_df[col].clip(lower=0)

share_sum = future_cov_df[energy_share_cols].sum(axis=1)
share_sum = share_sum.replace(0, np.nan)

for col in energy_share_cols:
    future_cov_df[col] = future_cov_df[col] / share_sum

future_cov_df['Year_dt'] = pd.to_datetime(future_cov_df['Year'].astype(int), format='%Y')

print(">>> 未来协变量预测完成")
print(future_cov_df.head())

# ========== 4. 拼接历史数据和未来协变量 ==========
future_for_model = future_cov_df.copy()
future_for_model[target_col] = np.nan

hist_for_model = df[['Province', 'Year', 'Year_dt', target_col] + cov_cols].copy()

full_df = pd.concat(
    [
        hist_for_model,
        future_for_model[['Province', 'Year', 'Year_dt', target_col] + cov_cols]
    ],
    ignore_index=True
)

full_df = full_df.sort_values(['Province', 'Year']).reset_index(drop=True)

# ========== 5. 转换为 Darts 对象 ==========
series = TimeSeries.from_group_dataframe(
    df,
    group_cols='Province',
    time_col='Year_dt',
    value_cols=target_col,
    freq='YS',
    fill_missing_dates=True
)

covariates_full = TimeSeries.from_group_dataframe(
    full_df,
    group_cols='Province',
    time_col='Year_dt',
    value_cols=cov_cols,
    freq='YS',
    fill_missing_dates=True
)

# ========== 6. 静态变量编码 + 标准化 ==========
static_transformer = StaticCovariatesTransformer()
series_encoded = static_transformer.fit_transform(series)

scaler_series = Scaler()
series_scaled = scaler_series.fit_transform(series_encoded)

scaler_cov = Scaler()
cov_scaled = scaler_cov.fit_transform(covariates_full)

# ========== 7. 定义 TFT 模型 ==========
model_tft = TFTModel(
    input_chunk_length=6,
    output_chunk_length=4,
    hidden_size=32,
    lstm_layers=1,
    num_attention_heads=4,
    dropout=0.2,
    batch_size=16,
    n_epochs=300,
    random_state=42,
    use_static_covariates=True,
    add_relative_index=True,
    pl_trainer_kwargs={
        "accelerator": "cpu",
        "enable_progress_bar": True
    }
)

# ========== 8. 训练 TFT ==========
print(">>> 正在训练 TFT 碳强度模型...")

model_tft.fit(
    series=series_scaled,
    past_covariates=cov_scaled,
    verbose=True
)

# ========== 9. 预测 2023-2026 ==========
print(">>> 正在预测 2023-2026 年碳排放强度...")

pred_scaled = model_tft.predict(
    n=4,
    series=series_scaled,
    past_covariates=cov_scaled,
    num_samples=1
)

pred = scaler_series.inverse_transform(pred_scaled)

# ========== 10. 整理碳强度预测结果 ==========
forecast_list = []

for idx in range(len(pred)):
    province_name = series[idx].static_covariates['Province'].iloc[0]

    pred_df = pred[idx].to_dataframe().reset_index()

    temp = pd.DataFrame()
    temp['Year'] = pred_df.iloc[:, 0].dt.year
    temp['Province'] = province_name
    temp['Carbon_Intensity_pred'] = pred_df.iloc[:, 1].values

    forecast_list.append(temp)

forecast_df = pd.concat(forecast_list, ignore_index=True)

# ========== 11. 合并能源结构预测 ==========
merge_cols = ['Province', 'Year'] + cov_cols

final_df = forecast_df.merge(
    future_cov_df[merge_cols],
    on=['Province', 'Year'],
    how='left'
)

final_df = final_df.sort_values(['Province', 'Year']).reset_index(drop=True)

# ========== 12. 导出 ==========
final_df.to_csv(
    'TFT_Carbon_Intensity_and_Energy_Structure_2023_2026.csv',
    index=False,
    encoding='gbk'
)

future_cov_df.to_csv(
    'Future_Energy_Structure_2023_2026.csv',
    index=False,
    encoding='gbk'
)

print(">>> 已导出：TFT_Carbon_Intensity_and_Energy_Structure_2023_2026.csv")
print(">>> 已导出：Future_Energy_Structure_2023_2026.csv")

# ========== 13. 全国平均趋势图 ==========
hist_avg = df.groupby('Year')[target_col].mean().reset_index()

forecast_avg = final_df.groupby('Year')['Carbon_Intensity_pred'].mean().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(
    hist_avg['Year'],
    hist_avg[target_col],
    marker='o',
    label='Historical Carbon Intensity'
)

plt.plot(
    forecast_avg['Year'],
    forecast_avg['Carbon_Intensity_pred'],
    marker='o',
    label='Forecast Carbon Intensity'
)

plt.axvline(x=2022, linestyle='--')
plt.xlabel('Year')
plt.ylabel('Carbon Intensity')
plt.title('Historical and Forecasted Carbon Intensity, 2000-2026')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(">>> 最终预测结果预览：")
print(final_df.head(20))
```

```python
# ========== 英文字体 + 16:9 3D 折线图 ==========
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
import numpy as np
import pandas as pd

# ===== 字体设置（英文）=====
matplotlib.rcParams['font.family'] = 'DejaVu Sans'
matplotlib.rcParams['axes.unicode_minus'] = False

# ========== 1. 数据准备 ==========
# 历史数据
hist_plot = df[['Province', 'Year', 'carbon_intensity']].copy()
hist_plot = hist_plot.rename(columns={'carbon_intensity': 'Value'})

# 预测数据
future_plot = final_df[['Province', 'Year', 'Carbon_Intensity_pred']].copy()
future_plot = future_plot.rename(columns={'Carbon_Intensity_pred': 'Value'})

# 合并
plot_df = pd.concat([hist_plot, future_plot], ignore_index=True)
plot_df = plot_df.sort_values(['Province', 'Year'])

# ========== 2. 按平均值排序 ==========
province_order = (
    plot_df.groupby('Province')['Value']
    .mean()
    .sort_values(ascending=False)
    .index
    .tolist()
)

province_to_x = {p: i for i, p in enumerate(province_order)}

# ========== 3. 创建 16:9 图 ==========
fig = plt.figure(figsize=(19.2, 10.8))
ax = fig.add_subplot(111, projection='3d')

# ========== 4. 绘制每个省份 ==========
for province in province_order:

    temp = plot_df[
        plot_df['Province'] == province
    ].sort_values('Year')

    x = np.full(len(temp), province_to_x[province])
    y = temp['Year'].values
    z = temp['Value'].values

    ax.plot(
        x,
        y,
        z,
        linewidth=2
    )

# ========== 5. 坐标轴 ==========
ax.set_xlabel('Province', labelpad=25)
ax.set_ylabel('Year', labelpad=15)
ax.set_zlabel('Carbon Intensity', labelpad=15)

ax.set_title(
    'Provincial Carbon Intensity Trends (Historical + Forecast)',
    fontsize=20,
    pad=25
)

# X轴省份标签
ax.set_xticks(range(len(province_order)))

ax.set_xticklabels(
    province_order,
    rotation=90,
    fontsize=7
)

# 年份
ax.set_yticks(
    sorted(plot_df['Year'].unique())[::3]
)

# ========== 6. 视角 ==========
ax.view_init(
    elev=18,
    azim=-68
)

# ========== 7. 去除背景面 ==========
ax.grid(False)

ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

# ========== 8. 调整布局 ==========
# 底部给标签留空间（不压缩主体16:9）
plt.subplots_adjust(
    left=0.03,
    right=0.98,
    top=0.92,
    bottom=0.28
)

plt.show()
```

```python
# ========== 区域异质性分析：历史 + 预测 ==========
import pandas as pd
import matplotlib.pyplot as plt

# 1. 省份区域划分
region_map = {
    # 东部
    'Beijing': 'East', 'Tianjin': 'East', 'Hebei': 'East',
    'Shanghai': 'East', 'Jiangsu': 'East', 'Zhejiang': 'East',
    'Fujian': 'East', 'Shandong': 'East', 'Guangdong': 'East',
    'Hainan': 'East',

    # 中部
    'Shanxi': 'Central', 'Anhui': 'Central', 'Jiangxi': 'Central',
    'Henan': 'Central', 'Hubei': 'Central', 'Hunan': 'Central',

    # 西部
    'Inner Mongolia': 'West', 'Guangxi': 'West', 'Chongqing': 'West',
    'Sichuan': 'West', 'Guizhou': 'West', 'Yunnan': 'West',
    'Shaanxi': 'West', 'Gansu': 'West', 'Qinghai': 'West',
    'Ningxia': 'West', 'Xinjiang': 'West',

    # 东北
    'Liaoning': 'Northeast', 'Jilin': 'Northeast', 'Heilongjiang': 'Northeast'
}

# 2. 历史数据
hist_region = df[['Province', 'Year', 'carbon_intensity']].copy()
hist_region = hist_region.rename(columns={'carbon_intensity': 'Carbon_Intensity'})
hist_region['Data_Type'] = 'Historical'

# 3. 预测数据
future_region = final_df[['Province', 'Year', 'Carbon_Intensity_pred']].copy()
future_region = future_region.rename(columns={'Carbon_Intensity_pred': 'Carbon_Intensity'})
future_region['Data_Type'] = 'Forecast'

# 4. 合并
region_df = pd.concat([hist_region, future_region], ignore_index=True)

# 5. 加区域标签
region_df['Region'] = region_df['Province'].map(region_map)

# 如果有没匹配上的省份，打印出来
missing_region = region_df[region_df['Region'].isna()]['Province'].unique()
print(">>> 未匹配区域的省份：", missing_region)

region_df = region_df.dropna(subset=['Region'])

# 6. 计算区域年度平均碳强度
region_avg = (
    region_df
    .groupby(['Region', 'Year'])['Carbon_Intensity']
    .mean()
    .reset_index()
)

# 7. 画区域异质性趋势图
plt.figure(figsize=(12, 6.75))  # 16:9

for region in region_avg['Region'].unique():
    temp = region_avg[region_avg['Region'] == region]
    plt.plot(
        temp['Year'],
        temp['Carbon_Intensity'],
        marker='o',
        linewidth=2,
        label=region
    )

plt.axvline(x=2022, linestyle='--', linewidth=1.5, label='Forecast Start')

plt.xlabel('Year')
plt.ylabel('Carbon Intensity')
plt.title('Regional Heterogeneity of Carbon Intensity, 2000-2026')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.savefig(
    'Regional_Heterogeneity_Carbon_Intensity.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()

print(">>> 已保存：Regional_Heterogeneity_Carbon_Intensity.png")

# 8. 导出区域平均结果
region_avg.to_csv(
    'Regional_Average_Carbon_Intensity_2000_2026.csv',
    index=False,
    encoding='gbk'
)

print(">>> 已导出：Regional_Average_Carbon_Intensity_2000_2026.csv")
print(region_avg.tail(20))
```
